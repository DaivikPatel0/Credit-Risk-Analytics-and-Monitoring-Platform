# Data Dictionary — Loan Default Prediction Dataset

## 1. Applicant / Identity & Admin Columns
| Column                  | Description                                                                                                     |
| ----------------------- | --------------------------------------------------------------------------------------------------------------- |
| **ID**                  | Unique identifier for each loan applicant / loan record. Used as primary key.                                   |
| **Batch Enrolled**      | Internal batch or cohort identifier representing when/where the loan was processed. Useful for cohort analysis. |
| **Application Type**    | Indicates whether the loan application is from an **INDIVIDUAL** or **JOINT** applicant.                        |
| **Verification Status** | Whether applicant information was verified by source, verified, or not verified.                                |
| **Initial List Status** | Loan listing status at origination (`w` = waiting, `f` = forwarded).                                            |
| **Payment Plan**        | Indicates special payment plan participation. Constant value (`n`) → low analytical value.                      |

## 2. Loan Characteristics
| Column                     | Description                                                                          |
| -------------------------- | ------------------------------------------------------------------------------------ |
| **Loan Amount**            | Total loan amount requested by the applicant.                                        |
| **Funded Amount**          | Actual loan amount funded/approved by the bank.                                      |
| **Funded Amount Investor** | Portion of the funded loan amount contributed by investors (fractional funding).     |
| **Term**                   | Loan repayment duration in months (e.g., 36, 58, 59).                                |
| **Interest Rate**          | Annual interest rate (%) applied to the loan.                                        |
| **Grade**                  | Bank-assigned credit grade (A–G), where A indicates lower risk.                      |
| **Sub Grade**              | More granular risk classification within Grade (e.g., B3, C1).                       |
| **Loan Title**             | Purpose or category of the loan (e.g., debt consolidation, credit card refinancing). |

## 3. Employment & Housing Profile
| Column                  | Description                                                                                                                   |
| ----------------------- | ----------------------------------------------------------------------------------------------------------------------------- |
| **Employment Duration** | Applicant’s employment/housing indicator (Mortgage, Rent, Own). Despite name, acts as housing/employment stability indicator. |
| **Home Ownership**      | Numeric indicator likely representing estimated housing value or ownership-related financial measure (not categorical).       |


## 4. Credit Behavior & Risk Indicators (Pre-Loan)
| Column                      | Description                                                            |
| --------------------------- | ---------------------------------------------------------------------- |
| **Debit to Income**         | Debt-to-income ratio of applicant (monthly debt / monthly income).     |
| **Delinquency - two years** | Number of times applicant was 30+ days delinquent in the past 2 years. |
| **Inquires - six months**   | Number of credit inquiries in the last 6 months.                       |
| **Open Account**            | Number of currently open credit accounts.                              |
| **Public Record**           | Count of derogatory public records (bankruptcies, liens, etc.).        |
| **Revolving Balance**       | Total outstanding revolving credit balance.                            |
| **Revolving Utilities**     | Revolving credit utilization percentage.                               |
| **Total Accounts**          | Total number of credit accounts (open + closed).                       |
| **Accounts Delinquent**     | Number of delinquent accounts (constant zero → low value).             |

## 5. Payment Performance & Collection Metrics (Post-Loan)
| Column                           | Description                                                                                    |
| -------------------------------- | ---------------------------------------------------------------------------------------------- |
| **Total Received Interest**      | Total interest amount received by lender to date.                                              |
| **Total Received Late Fee**      | Total late payment fees collected.                                                             |
| **Recoveries**                   | Amount recovered after loan charge-off.                                                        |
| **Collection Recovery Fee**      | Fees charged for recovery efforts after charge-off.                                            |
| **Collection 12 months Medical** | Indicator of medical collections in past 12 months.                                            |
| **Total Collection Amount**      | Total amount ever sent to collections.                                                         |
| **Total Current Balance**        | Current outstanding balance across all borrower accounts.                                      |
| **Total Revolving Credit Limit** | Total credit limit available to the borrower on revolving accounts.                            |
| **Last week Pay**                | Number of weeks borrower continued payments after enrollment. Proxy for repayment consistency. |

## 6. Target Variable
| Column          | Description                                                                    |
| --------------- | ------------------------------------------------------------------------------ |
| **Loan Status** | **Target variable** — `1 = Defaulter`, `0 = Non-Defaulter`. Highly imbalanced. |
