```python
import os
import pandas as pd
from dotenv import load_dotenv
from sqlalchemy import create_engine
```

```python
load_dotenv("C:/Users/daivi/Desktop/bank-credit-risk-platform/.env.txt") 

engine = create_engine(
    f"postgresql+psycopg2://{os.getenv('PG_USER')}:{os.getenv('PG_PASSWORD')}@"
    f"{os.getenv('PG_HOST')}:{os.getenv('PG_PORT')}/{os.getenv('PG_DB')}"
)
```

```python
with engine.connect() as conn:
    print("Connected to PostgreSQL")
```

```python
df = pd.read_csv("C:/Users/daivi/Desktop/bank-credit-risk-platform/data/Processed/loan_applications_clean.csv")

df.to_sql(
    "loan_applications",
    engine,
    if_exists="replace",
    index=False
)

print("Table loaded: loan_applications")
```

```python
pd.read_sql("SELECT COUNT(*) AS rows FROM loan_applications;", engine)
```
